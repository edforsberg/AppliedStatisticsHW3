```python
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_validate
import statistics
import pandas as pd
```

```python
X, y = load_boston(return_X_y=True)
median = statistics.median(y)
y = [int(x>median) for x in y]
nr_folds = 10
```

```python
##logistic regression

import sklearn.linear_model as lm

logistic_regression = lm.LogisticRegression(solver = 'liblinear', warm_start = True)
cv_results_lr = cross_validate(logistic_regression, X, y, cv=nr_folds, error_score='raise', return_train_score=True)

df_lr = pd.DataFrame(data=cv_results_lr)
df_lr.loc['mean'] = df_lr.mean()
df_lr
```

```python
##LDA 

import sklearn.discriminant_analysis as dm

linear_discriminant_analysis = dm.LinearDiscriminantAnalysis(solver = 'svd') #'lsqr', 'eigen'

cv_results_lda = cross_validate(linear_discriminant_analysis, X, y, cv=nr_folds, error_score='raise', return_train_score=True)

df_lda = pd.DataFrame(data=cv_results_lda)
df_lda.loc['mean'] = df_lda.mean()
df_lda
```

```python
##KNN

import sklearn.neighbors as knn

k_nearest_neighbours = knn.KNeighborsClassifier(n_neighbors = 10) #biased?

cv_results_knn = cross_validate(k_nearest_neighbours, X, y, cv=nr_folds, error_score='raise', return_train_score=True)

df_knn = pd.DataFrame(data=cv_results_knn)
df_knn.loc['mean'] = df_knn.mean()
df_knn
```

```python
##decision tree
import sklearn.tree as tree

decision_tree = tree.DecisionTreeClassifier(min_samples_split = 2, min_samples_leaf = 2)

cv_results_dt = cross_validate(decision_tree, X, y, cv=nr_folds, error_score='raise', return_train_score=True)
bagging_results_dt = ens.BaggingClassifier(decision_tree, n_estimators = 100)

df_dt = pd.DataFrame(data=cv_results_dt)
df_dt.loc['mean'] = df_dt.mean()
df_dt
```

```python
##decision tree with bagging
import sklearn.tree as tree
import sklearn.ensemble as ens

bagging_decision_tree = ens.BaggingClassifier(base_estimator = tree.DecisionTreeClassifier(min_samples_split = 2, min_samples_leaf = 2))

cv_results_dt_bagging = cross_validate(bagging_decision_tree, X, y, cv=nr_folds, error_score='raise', return_train_score=True)

df_dt_bagging = pd.DataFrame(data=cv_results_dt_bagging)
df_dt_bagging.loc['mean'] = df_dt_bagging.mean()
df_dt_bagging
```

```python
##Random forest classifier
import sklearn.ensemble as ens

random_forrest_classifier = ens.RandomForestClassifier(n_estimators = 100)

cv_results_rfc = cross_validate(random_forrest_classifier, X, y, cv=nr_folds, error_score='raise', return_train_score=True)

df_rfc = pd.DataFrame(data=cv_results_rfc)
df_rfc.loc['mean'] = df_rfc.mean()
df_rfc
```
